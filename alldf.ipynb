```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.util import *
from textblob import TextBlob
from nltk import tokenize
```

```python
alluni.drop_duplicates(subset ="Review Text", keep = "first", inplace = True)
alluni['Review Text'] = alluni['Review Text'].astype('str')
def get_polarity(text):
    return TextBlob(text).sentiment.polarity
alluni['Polarity'] = alluni['Review Text'].apply(get_polarity)
```

```python
alluni['Sentiment_Type']=''
alluni.loc[alluni.Polarity>0,'Sentiment_Type']='POSITIVE'
alluni.loc[alluni.Polarity==0,'Sentiment_Type']='NEUTRAL'
alluni.loc[alluni.Polarity<0,'Sentiment_Type']='NEGATIVE'
```

```python
alluni
```

```python
alluni.drop_duplicates(subset ="Review Text", keep = "first", inplace = True)
alluni['Review Text'] = alluni['Review Text'].astype('str')
def get_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity
alluni['subjectivity'] = alluni['Review Text'].apply(get_subjectivity)
```
